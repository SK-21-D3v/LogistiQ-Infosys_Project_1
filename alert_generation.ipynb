```python
import pandas as pd
import pandas as pd
import numpy as np
# Load the existing electronics data
file_path = 'Final_product_dataset.csv'
data = pd.read_csv(file_path)
df = pd.DataFrame(data)

# Define thresholds for alerts
risk_threshold = 5
inventory_threshold = 20
supplier_threshold = 4
transportation_threshold = 8
negative_sentiment = 'Negative'

# Create alerts
alerts = []

# Check conditions and create alerts
for index, row in df.iterrows():
    if row['Risk Factor'] > risk_threshold:
        alerts.append(f"Alert: High risk factor detected for product '{row['product_name']}' (Risk Factor: {row['Risk Factor']})")
    if row['Inventory Level (%)'] < inventory_threshold:
        alerts.append(f"Alert: Low inventory for product '{row['product_name']}' (Inventory Level: {row['Inventory Level (%)']}%)")
    if row['Supplier Reliability'] < supplier_threshold:
        alerts.append(f"Alert: Low supplier reliability for product '{row['product_name']}' (Reliability: {row['Supplier Reliability']})")
    if row['Transportation Risk'] > transportation_threshold:
        alerts.append(f"Alert: High transportation risk for product '{row['product_name']}' (Transportation Risk: {row['Transportation Risk']})")

# Display alerts
for alert in alerts:
    print(alert)
```
